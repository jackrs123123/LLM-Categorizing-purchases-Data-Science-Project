```python
from langchain_community.llms import Ollama
```

```python
llm = Ollama(model='llama2')
```

```python
import pandas as pd
df = pd.read_csv('data.csv', encoding='unicode_escape')
df.head()
```

```python
unique_items = df['Description'].unique()
print(unique_items, '\n', len(unique_items), '<---- is total amount of items!')
```

```python
def hop(start, stop, step):
    for i in range(start, stop, step):
        yield i
    yield stop

index_list = list(hop(0, len(unique_items), 30))
index_list
```

```python
def categorize(item_names, llm):
    request = llm.invoke('Please categorize each single purchase. Examples: Lantern/Light - Decorum, Block/Doll/Playhouse -  Toys, Coat/Fashion - Clothing, Food/Water - Groceries.' + ','.join(unique_items[0:30]))
    request = request.split('\n')

    categorize_df = pd.Series()
    for i in range(len(item_names)+2):
        categorize_df.loc[len(categorize_df)] = request[i]

    return categorize_df
```
